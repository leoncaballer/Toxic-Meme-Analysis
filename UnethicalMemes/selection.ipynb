```python
import json

# Replace 'file_path.jsonl' with the path to your file
file_path = "/workspaces/Toxic-Meme-Analysis/UnethicalMemes/annotation/train.json"

with open(file_path, 'r') as file:
    for line in file:
        try:
            # Parse the line as JSON
            data = json.loads(line.strip())
            
            # Access specific keys
            id_ = data.get("id")
            img = data.get("img")
            labels = data.get("labels")
            text = data.get("text")
            caption = data.get("caption")
            
            # Example: Print the parsed data
            print(f"ID: {id_}, Image: {img}, Labels: {labels}, Text: {text}, Caption: {caption}")
        
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON: {e}")
```

```python
# List to hold entries with label 0
label_0_entries = []

with open(file_path, 'r') as file:
    for line in file:
        try:
            # Clean up the line by removing extra spaces and newlines
            line = line.strip()

            # Try parsing the line as JSON
            data = json.loads(line)

            # Check if the label is 0
            if 0 in data.get("labels", []):
                label_0_entries.append(data)

        except json.JSONDecodeError as e:
            # Print the error and the line where it occurred
            print(f"Error decoding JSON: {e} in line: {line}...")  # Print first 100 chars of line for context

# After the loop, 'label_0_entries' contains the first 500 entries with label 0
print(f"Extracted {len(label_0_entries)} entries with label 0")

with open('harmless_entries.jsonl', 'w') as output_file:
    for entry in label_0_entries:
        json.dump(entry, output_file)
        output_file.write('\n')
        
```

```python
import os
import json
import shutil

# File paths
metadata_file = "/workspaces/Toxic-Meme-Analysis/UnethicalMemes/harmless_entries_update.json"  # Path to the harmless entries file
image_folder = "/workspaces/Toxic-Meme-Analysis/UnethicalMemes/images"  # Path to the folder containing the images
output_folder = "/workspaces/Toxic-Meme-Analysis/ImageClustering/harmless_images"  # Path to the folder where images will be copied

# Load metadata from the JSON file
with open(metadata_file, 'r') as file:
    metadata = json.load(file)

# Select the first 500 image entries from the metadata (or any specific set of IDs)
# If you have a specific list of IDs to select, you can define them here as a list
selected_ids = [entry['id'] for entry in metadata[:500]]  # Select first 500 images

# Make sure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Iterate over the metadata and copy the corresponding images to the output folder
for entry in metadata:
    if entry['id'] in selected_ids:
        img_file = entry['img']
        img_path = os.path.join(image_folder, img_file)
        
        # Check if the image exists in the folder
        if os.path.exists(img_path):
            # Copy the image to the output folder
            shutil.copy(img_path, os.path.join(output_folder, img_file))

print(f'Successfully copied 500 selected images to {output_folder}')
```
